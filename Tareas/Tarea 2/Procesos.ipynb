Realiza el Programa 1 que instancie, como clase o con el método Process, 10 procesos. Cada uno de los procesos hijos recibirán un valor entero y un caracter $(i,c)$ enviados por el proceso padre, los procesos escribirán en la salida estándar $i$ veces el caracter $c$.

```
import multiprocessing as mp
import random as rd


def alfanum(i,c):
    print("Proceso con caracter "+ str(c)+ " con " + str(i)+" veces")
    for _ in range(i):
        print(c)


caracteres=['a','b','c','d','e','f','g','h','i','j','k','l','m','o','p','q','r','s','t','u','v','w','x','y','z']
repeticiones=rd.sample(range(1,11),10)

if __name__ == '__main__':
    lista = []
    for i in repeticiones:
        char= rd.choice(caracteres)
        p = mp.Process(target=alfanum, args=(i,char ,))
        lista.append(p)
    for i in lista:
        i.start()
    for i in lista:
        i.join()
```


2. Refactoriza (reescribe) el programa anterior y elabora el Programa2 que incluya un mecanismo de sin- cronización el cual permita escribir en orden todos los caracteres de cada proceso. Es decir, que se obtenga la secuencia $c_{1,1}...c_{1,i}, c_{2,1}...c_{2,i}, ..., c_{10,1}...c_{10,i}$ donde cada subsecuencia $c_{k,i}$ para cada $k = 1, 2, ..., 10$ es la secuencia de caracteres del proceso hijo k con longitud $i$.  
Hint: Prueba usar turnos.

```
import multiprocessing as mp
import random as rd
import time

lista = []

def alfanum(i, c, lock):
    global lista
    lock.acquire()
    print("Proceso con caracter "+ str(c)+ " con " + str(i)+" veces")
    time.sleep(.1)
    for _ in range(i):
        print(c)
        lista.append(c)
    lock.release()
    

caracteres = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'o', 'p',
              'q', 'r', 's', 't', 'u', 'v',
              'w', 'x', 'y', 'z']
repeticiones=rd.sample(range(1,11),10)

if __name__ == '__main__':
    lock=mp.Lock()
    for i in repeticiones:
        char= rd.choice(caracteres)
        p = mp.Process(target=alfanum, args=(i,char , lock))
        lista.append(p)
    for i in lista:
        i.start()
    for i in lista:
        i.join()
```
