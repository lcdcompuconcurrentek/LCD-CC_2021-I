```python
import threading
from collections import deque
import time

tama単o_almacen = 4
q = deque([])
n = 0 

lock=threading.Lock()
condicion_productor = threading.Condition(lock)
condicion_consumidor = threading.Condition(lock)    

class Productor:
    def run(self):
        global tama単o_almacen, q, n
        while True:
            with condicion_productor:
                while len(q) >= tama単o_almacen:
                    print( "El alamacen esta lleno. Elementos en almacen = ", len(q))
                    condicion_consumidor.notify()
                    condicion_productor.wait()    

                q.append(n)
                print("Producido ", n)
                n = n + 1
                condicion_consumidor.notify()


class Consumidor:
    def run(self):
        global tama単o_almacen, q, n
        while True:
            with condicion_consumidor:
                while len(q) == 0:
                    print( "El almacen esta vacio. Elementos en almacen =", len(q) )
                    condicion_productor.notify() 
                    condicion_consumidor.wait()   

                x = q.popleft()
                print("Consumido ", x)
                time.sleep(1)
                condicion_productor.notify()   

p = Productor()
c = Consumidor()
pThread = threading.Thread( target=p.run, args=())
cThread = threading.Thread( target=c.run, args=())
pThread.start()
cThread.start()
pThread.join()
cThread.join()
```
